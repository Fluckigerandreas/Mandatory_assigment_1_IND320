```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Read CSV
df = pd.read_csv("open-meteo-subset.csv")

# 2. Print info
print(df.info())
print(df.head())

# 3. Plot each column separately
for column in df.columns:
    if pd.api.types.is_numeric_dtype(df[column]):
        plt.figure(figsize=(8, 4))
        plt.plot(df[column])
        plt.title(f"{column} over time")
        plt.xlabel("Index")
        plt.ylabel(column)
        plt.grid(True)
        plt.show()

# 4. Plot all numeric columns together (normalized)
numeric_cols = df.select_dtypes(include='number').columns
df_numeric = df[numeric_cols]
df_normalized = (df_numeric - df_numeric.min()) / (df_numeric.max() - df_numeric.min())

plt.figure(figsize=(10, 6))
for column in df_normalized.columns:
    plt.plot(df_normalized[column], label=column)

plt.title("All numeric columns (normalized to [0,1])")
plt.xlabel("Index")
plt.ylabel("Normalized values")
plt.legend()
plt.grid(True)
plt.show()
```
